```python
def get_playlist_tracks(username, playlist_id):
    
    results = sp.user_playlist_tracks(username, playlist_id)
    tracks = results['items']
    
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    
    return tracks

tracks = get_playlist_tracks(username,playlist_id)
```

# Starting with Kmeans unsupervised learning

```python
playlist_for_clustering = df_spotify_features[[
                             "danceability", "energy", "key", "loudness", "mode", "speechiness","acousticness",
                             "instrumentalness", "liveness", "valence", "tempo", "duration_ms", "time_signature"]]
```

```python

# K-Means is a distance based algorithm: we need to scale / normalize:
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_prep = scaler.fit_transform(playlist_for_clustering)

pd.DataFrame(X_prep).head()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=10, random_state=1234)

kmeans.fit(X_prep)

kmeans.get_params()
```

```python
# Predicting / assigning the clusters:
clusters = kmeans.predict(X_prep)
clusters
```

```python
# Check the size of the clusters
pd.Series(clusters).value_counts().sort_index()
```

```python
# Explore the cluster assignment in the original dataset
X_df = pd.DataFrame(playlist_for_clustering)
X_df["cluster"] = clusters
X_df.head()
```

```python
X_df
```

```python
df1['cluster'] = X_df['cluster']
```
